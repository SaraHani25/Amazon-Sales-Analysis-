```python
## Are there significant outliers in product prices, discounts, and ratings that may affect analysis?
columns_to_plot = ['discounted_price', 'actual_price', 'discount_percentage', 'rating', 'rating_count']
fig, axs = plt.subplots(ncols=5, nrows=1, figsize=(25, 6))
axs = axs.flatten()
for index, k in enumerate(columns_to_plot):
    sns.boxplot(y=k, data=df, ax=axs[index])
plt.tight_layout(pad=0.4, w_pad=0.5, h_pad=5.0)
plt.show()
```

#### Boxplots show clear outliers in prices and rating counts, while ratings stay more consistent.
